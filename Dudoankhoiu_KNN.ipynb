**Yêu cầu 2:**
Thông kê dữ liệu missing cho từng thuộc tính, và liệt kê ra các bản ghi bị missing tương ứng với thuộc tính đó; suy nghĩ phương án xử lý phù hợp

```python
def missing_values(df):
        mis_val = df.isnull().sum()
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Số giá trị Missing', 1 : 'Tỷ lệ % missing'})
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        'Tỷ lệ % missing', ascending=False).round(2)
        print ("File dữ liệu bao gồm có: " + str(df.shape[1]) + " cột.\n"      
            "Có " + str(mis_val_table_ren_columns.shape[0]) +
              " cột chứa missing values.")
        return mis_val_table_ren_columns
```

# B1.3.3: Chia tập dữ liệu thành 2 phần Train-test với tỷ lệ 75-25

### 3. Tách tập dữ liệu ra tập Training - Test: 
---
chia dữ liệu thành 2 phần Train-Test tỷ lệ là 75%-25%

```python
#Gọi hàm tách dữ liệu trong thư viện sklearn:
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_Practice,
                                                   Y_Practice,
                                                   #train_size= 0.75,
                                                    test_size = 0.25,
                                                   shuffle=True,
                                                   random_state=42) #(shuffle=True |shuffle=fail lấy 80% ngẫu nhiên| lấy 80% từ trên xuống theo ds)


print("1. Kích thước tập Training: %d" %len(Y_train))
print("2. Kích thước tập Test    : %d" %len(Y_test))
```

```python
# Hiển thị X_test
print(X_test)
print(X_test.shape)
```

```python
# Hiển thị Y_test
print(Y_test)
print(Y_test.shape)
```

# B1.4: Xây dựng mô hình KNN

Xây dựng mô hình KNN 

```python
#Gọi thư viện sklearn với thuật toán phân lớp KNN
from sklearn.neighbors import KNeighborsClassifier

# #Khai báo model KNN sử dụng các tham số mặc định
# KNN_Practice = KNeighborsClassifier(n_neighbors= 15, #Số người láng giêng
#                                   p =2) # Phương pháp tính khoảng cách
KNN_Practice = KNeighborsClassifier(n_neighbors=11,weights='distance',p=2)

#Hiển thị danh sách tham số:
KNN_Practice.get_params()
```

```python
#huấn luyện mô hình
KNN_Practice.fit(X_train, Y_train)
```

# B1.a: Độ chính xác của model trên tập Train - Test

```python
#Độ chính xác của mô hình huấn luyện
acc_train = KNN_Practice.score(X_train,Y_train)
print('Độ chính xác trên tập huấn luyện:', round(acc_train*100,2))
```

```python
#Sử dụng mô hình huấn luyện dự đoán trên tập dữ liệu Test
Y_pred = KNN_Practice.predict(X_test)

#Kết quả dự đoán của mô hình trên tập test
print(Y_pred)
```

```python
print("Kết quả dự đoán khối u ác tính hay lành tính trên tập dữ liệu TEST:")
print("Kết quả cho 20 bệnh nhân đầu tiên:")
print("Predicted labels: ",Y_pred[:20] )
print("Ground truth    : ",Y_test[:20] )
```

# B1.b: Tổng số mẫu dự đoán đúng trên tập Test

```python
#Sử dụng module accuracy_score trong thư viện sklearn để đánh giá độ chính xác:
from sklearn.metrics import accuracy_score

#1. Đếm tổng số mẫu dự đoán đúng trên tập Test
acc1 = accuracy_score(Y_test,Y_pred,
                     normalize=False) # Nếu normalize=False là đếm số lượng
print('Tổng số mẫu dự đoán trên tập test: ', acc1,'/', len(Y_test))
```

```python
#2.Độ chính xác theo tỷ lệ % số mẫu dự đoán đúng/tổng số mẫu của tập test (Accuracy)
acc2 =  accuracy_score(Y_test,Y_pred,
                     normalize=True) # Nếu normalize=True là tính tỷ lệ %
print('Tổng số mẫu dự đoán trên tập test: ', acc2)
print( 'Tỷ lệ % :',round (acc2*100,2),'%')
```

### 4. Xây dựng mô hình KNN cho bài toán phân lớp
---
Khai báo model KNN sử dụng thư viện sklearn

```python
#Gọi thư viện sklearn với thuật toán phân lớp KNN
from sklearn.neighbors import KNeighborsClassifier

# #Khai báo model KNN sử dụng các tham số mặc định
# KNN_Practice = KNeighborsClassifier(n_neighbors= 5, #Số người láng giêng
#                                   p =2) # Phương pháp tính khoảng cách
KNN_Practice = KNeighborsClassifier(n_neighbors=9,weights='distance',p=2)

#Hiển thị danh sách tham số:
KNN_Practice.get_params()
```

### Một số tham số quan trọng của thuật toán KNN trong thư viện sklearn:
Tham khảo KNN của thư viện sklearn:
https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html


1. n_neighbors: số điểm lân cận sử dụng (default = 5)
2. weights: có tính trọng số của các điểm lân cận hay không?
    * uniform: Tất cả các điểm lân cận đều có trọng số như nhau (default)
    * distance: Các điểm lân cận gần hơn sẽ được đánh trọng số cao hơn
3. algorithm: Thuật toán sử dụng để tính các điểm lân cận: ball_tree, kd_tree, brute, auto (default)
4. p: xác định khoảng cách tính theo manhattan (p=1) hay tính theo euclidean (p=2 default)

Trong thực tế ta thấy rằng những điểm gần hơn nên có trọng số cao hơn (càng thân cận thì càng tin tưởng). Vậy nên sẽ đánh trọng số khác nhau cho mỗi trong 11 điểm gần nhất này.
